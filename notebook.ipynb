```python
# Function to calculate the distance to the nearest road in meters
def get_distance_to_road_meters(geom):
    # Find the nearest edge and its distance
    nearest = ox.distance.nearest_edges(G_proj, geom.x, geom.y, return_dist=True)
    # The distance is the second element in the tuple
    return nearest[1]  # Now in meters since we're using a projected CRS

# Apply the function to calculate the distance to the nearest road for each branch
gdf_proj['distance_to_road_m'] = gdf_proj['geometry'].apply(get_distance_to_road_meters)
```

```python
# Calculate node degree (number of connected roads at each intersection)
node_degrees = dict(G.degree())

# Add degree of nearest node to each property
def get_node_degree(geom):
    nearest_node = ox.distance.nearest_nodes(G, geom.x, geom.y)
    return node_degrees[nearest_node]

final_df['node_degree'] = final_df['geometry'].apply(get_node_degree)
```

```python
import geopandas as gpd
from shapely.geometry import LineString, Point


# Convert network to GeoDataFrame of edges
edges = []
for u, v, key, data in G.edges(keys=True, data=True):
    if 'geometry' in data:
        # If the edge has a geometry attribute
        geom = data['geometry']
    else:
        # Create a line from node coordinates
        from_point = Point((G.nodes[u]['x'], G.nodes[u]['y']))
        to_point = Point((G.nodes[v]['x'], G.nodes[v]['y']))
        geom = LineString([from_point, to_point])
    
    # Add length if not present
    if 'length' not in data:
        data['length'] = geom.length
        
    edges.append((u, v, key, data, geom))

# Create GeoDataFrame
gdf_edges = gpd.GeoDataFrame(
    [{'u': u, 'v': v, 'key': k, 'length': d['length']} for u, v, k, d, _ in edges],
    geometry=[g for _, _, _, _, g in edges]
)
gdf_edges.crs = G.graph['crs']

# Calculate road density for each property
def calculate_road_density(geom, buffer_distance=500):
    """Calculate road density within a buffer area around a property"""
    # Create buffer
    buffer = geom.buffer(buffer_distance)
    
    # Get roads within buffer
    roads_in_buffer = gdf_edges[gdf_edges.intersects(buffer)]
    
    # Calculate total road length (in meters)
    total_road_length = roads_in_buffer['length'].sum()
    
    # Calculate buffer area (in square meters)
    buffer_area = buffer.area
    
    # Calculate density (meters of road per square meter)
    density = total_road_length / buffer_area if buffer_area > 0 else 0
    
    # Convert to km/kmÂ² for easier interpretation
    density_km = density * 1000000  # roads in km per kmÂ²
    
    return density_km

# Apply to each property
final_df['road_density_500m'] = final_df['geometry'].apply(
    lambda x: calculate_road_density(x, buffer_distance=500)
)
```

```python
# Save the enriched dataset with road network features
final_df.to_file("chicken_rep_lagos_branch.geojson", driver="GeoJSON")
```

```python
import osmnx as ox
import pandas as pd
import numpy as np
from shapely.geometry import Point
from geopy.distance import geodesic
```

```python
def get_nearby_pois(lat, lon, tags, radius=2000):
    try:
        point = (lat, lon)
        north, south, east, west = ox.utils_geo.bbox_from_point(point, dist=radius)
        bbox = (north, south, east, west)
        pois = ox.features.features_from_bbox(bbox, tags)
        pois['distance'] = pois.apply(
            lambda row: geodesic(point, (row['geometry'].centroid.y, row['geometry'].centroid.x)).meters,
            axis=1
        )
        pois = pois[pois['distance'] <= radius]
        return pois
    except Exception as e:
        print(f"Error for location ({lat}, {lon}): {e}")
        return pd.DataFrame()  # Return empty DataFrame
```

```python
tags = {
    'shop': ['mall', 'supermarket', 'market', 'convenience', 'department_store', 'grocery'],
    'office': ['company', 'coworking', 'government'],
    'amenity': ['marketplace', 'post_office', 'bank', 'atm', 'school', 'university', 'hospital', 'clinic'],
    'building': ['retail', 'commercial', 'office', 'civic'],
    'landuse': ['commercial', 'retail']
}
```

```python
all_pois = []
for index, row in final_df.iterrows():
    lat = row['Latitude']
    lon = row['Longitude']
    try:
        pois = get_nearby_pois(lat, lon, tags)
        if not pois.empty:
            pois['branch_location'] = row['Location']
            pois['branch_latitude'] = lat
            pois['branch_longitude'] = lon
            all_pois.append(pois)
            print(f"Success for location: {row['Location']}")
        else:
            print(f"No POIs found for location: {row['Location']}")
    except Exception as e:
        print(f"Error for location {row['Location']}: {e}")
```

```python
# Concatenate all POIs into a single DataFrame
nearby_pois_df = pd.concat(all_pois, ignore_index=True)
```

```python
# Create Point geometries from the Latitude and Longitude columns
geometry = [Point(xy) for xy in zip(nearby_pois_df['Longitude'], nearby_pois_df['Latitude'])]

# Convert the DataFrame to a GeoDataFrame
nearby_pois_gdf = gpd.GeoDataFrame(nearby_pois_df, geometry=geometry, crs="EPSG:4326")  
```
